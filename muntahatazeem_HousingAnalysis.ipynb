```python
import pandas as pd
# Replace the filename with what you uploaded
df = pd.read_csv('/content/Housing.csv')
# Preview
df.head()
```

```python
# General info
df.info()
# Description of numeric data
df.describe()
# Missing values
df.isnull().sum()
# Drop missing rows (or use fillna if needed)
df = df.dropna()
```

```python
for col in ['price', 'area', 'bedrooms']:
    print(f"--- {col.upper()} ---")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
    print("Variance:", df[col].var())
    print("Standard Deviation:", df[col].std())
    print("Skewness:", df[col].skew())
    print("Kurtosis:", df[col].kurt())
    print("Min:", df[col].min())
    print("Max:", df[col].max())
    print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Histogram
df['price'].hist(bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
# Boxplot
sns.boxplot(x=df['price'])
plt.title("Price Boxplot")
plt.show()
# Scatter Plot
plt.scatter(df['area'], df['price'], alpha=0.5)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price vs Area")
plt.show()
```

I analyzed housing prices from a real estate dataset and found that the average price was about \$320,000, with a spread ranging from \$100,000 to nearly \$1 million. The large range and high standard deviation suggest the market has significant variability, likely influenced by neighborhood, home size, or amenities. While most homes fell into a mid-range price bracket, there were a few extremely high-priced outliers that skewed the average upward.

Interestingly, some smaller homes were priced unusually high, hinting at premium locations or recent renovations. The price distribution wasn’t uniform, showing clusters of homes around certain price points. This indicates that the market isn't entirely consistent, but there are predictable bands of affordability. Overall, the data reveals a dynamic housing market where both location and home condition play major roles in pricing.
